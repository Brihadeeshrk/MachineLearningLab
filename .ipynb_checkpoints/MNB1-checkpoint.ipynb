```python
class MultiNomialNaiveBayesClassifier:
    def __init__(self, F, T):
        '''
        This is the class initializer
        Here, F are the features and T are the targets
        '''
        self.F, self.T = F, T
        self.N = len(self.F) # Length of Training Set
        self.dimensions = len(self.F[0]) # Dimension of Vector of Features
        self.attributes = [[] for _ in range(self.dimensions)] # storing the columns of the dataset
        self.output_classes = {} # o/p classes with the number of occurences in the training set; here we will have multiple classes
        self.data = [] # to store every row
        
        for i in range(len(self.F)):
            for j in range(len(self.dimensions));
            # if we have never seen this value for this attribute before, we add it to the attribute array in the
            # corresponding position
                if not self.F[i][j] in self.attributes[j]:
                    self.attributes[j].append(self.F[i][j])
                
            # if we have never seen this op class before, we add it to the output_classes and count one occurence for now
            if not self.T[i] in self.output_classes.keys():
                self.output_classes[self.T[i]] = 1
            else:
                self.output_classes(self.T[i]) += 1
                
            self.data.append([self.F[i], self.T[i]])
```
