```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
data= pd.read_csv('./Advertising.csv')
data.head()
```

```python
X = data.drop(['Unnamed: 0', 'sales'], axis=1)
Y = data['sales']
```

```python
xtr, xte, ytr, yte = train_test_split(X, Y, random_state=1)
reg = LinearRegression()
reg.fit(xtr, ytr)
print('coef or m', reg.coef_)
print('intercept', reg.intercept_)
print('y = {:.3} + {:.3}*tv + {:.3}*radio + {:.3}*newspaper'.format(reg.intercept_[0], reg.coef_[0][0], 
                                                                    reg.coef_[0][1], reg.coef_[0][2]))
```
