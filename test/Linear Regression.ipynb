```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('./Advertising.csv')
data.head()
data.columns
```

```python
data.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
myplt = plt.figure(figsize=(15, 10))
ax1 = myplt.add_subplot(2, 2, 1)
ax1.scatter(data['TV'], data['sales'], c='orange')
ax1.xlabel='money spent on tv ads'
ax1.ylabel='sales'

ax2 = myplt.add_subplot(2, 2, 2)
ax2.scatter(data['radio'], data['sales'], c='green')
ax2.xlabel='money spent on radio ads'
ax2.ylabel='sales'

ax3 = myplt.add_subplot(2, 2, 3)
ax3.scatter(data['newspaper'], data['sales'], c='brown')
ax3.xlabel='money spent on newspaper ads'
ax3.ylabel='sales'
```

```python
X = data['TV'].values.reshape(-1, 1)
Y = data['sales'].values.reshape(-1, 1)
xt, xte, yt, yte = train_test_split(X, Y, test_size=0.33)
reg = LinearRegression()
reg.fit(xt, yt)

print('slope or coeff: ', reg.coef_[0][0])
print('intercept: ', reg.intercept_[0])
print('y = {:.3}x + {:.3}'.format(reg.coef_[0][0], reg.intercept_[0]))
```

```python
pred = reg.predict(xte)
plt.figure(figsize=(15, 10))
plt.scatter(xte, yte, c='blue')
plt.plot(xte, pred, linewidth=2, c='orange')
plt.show()
```

```python
rmse = np.sqrt(mean_squared_error(yte, pred))
rmse
```
